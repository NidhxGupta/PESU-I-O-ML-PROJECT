```python
# Importing the libraries
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
#Importing the dataset
dataset = pd.read_csv('Iphone_Advertisement1.csv')
#Information was collected through a survey on how many bought an Iphone
```

```python
#Checking if Dataset is imported
dataset
```

```python
#Assigning the above values to x and y
x = dataset.iloc[:, [2,3]].values #Choosing Age and Estimated Salary as the Independent Variable x
y = dataset.iloc[:, 4].values #Choosing Purchased values (0/1) as Dependent Variable y
```

```python
#Checking if x and y are taken from dataset as required
x,y
```

```python
#Splitting the dataset into Training set and Test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0) #random state feeds the data as it is currently
#test_size decides the size of the data that has to be split as the test dataset
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Fitting Logistic Regression Model to the Training set
#Since it's a linear model, the customers would be divided into 2 categories
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)
y_pred
#y_pred is the vector of prediction i.e., it will give prediction for each of the test set observation
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
# A confusion matrix is a table that is often used to describe the performance of a classification model on a set of 
# test data for which the true values are known.
# As seen in the table, 44+22 = 66 is the total amount of correct predictions we have.
# And 5+2 = 7 is the total amount of incorrect predictions we have.
# (66/73)*100 = 90.41% accuracy
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('blue', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Visualising the Test set results
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('blue', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
